### Now let's load data from Directory Folder

```python
from langchain_classic.document_loaders.csv_loader import CSVLoader

loader = CSVLoader(file_path='./Directory/codebasics_faqs.csv', source_column="prompt")

# Store the loaded data in the 'data' variable
docs = loader.load()
```

### Hugging Face Embeddings Instruct Embeddings

```python
from langchain_classic.embeddings import HuggingFaceInstructEmbeddings
```

```python
instructor_embeddings = HuggingFaceInstructEmbeddings(query_instruction="Represent the query for  retrieval")

e = instructor_embeddings.embed_query("What is your refund policy?")
```

### Vector store using FAISS

```python
from langchain_classic.vectorstores import FAISS
instructor_embeddings = HuggingFaceInstructEmbeddings()
# In Case if I want to to Hugging Face Embeddings
# embeddings = HuggingFaceEmbeddings()
# Create a FAISS instance for vector database from 'data'
vectordb = FAISS.from_documents(documents=docs,
                                 embedding=instructor_embeddings)

# Create a retriever for querying the vector database
retriever = vectordb.as_retriever(score_threshold = 0.7)
```

### We are all set 👍🏼 Let's ask some questions now

```python
import os
from langchain_openai import ChatOpenAI
from langchain_classic.prompts import PromptTemplate
from langchain_classic.chains import RetrievalQA

llm = ChatOpenAI(
    model="gpt-4o-mini",
    temperature=0.1,
    api_key=os.environ["OPENAI_API_KEY"]
)

prompt_template = """Given the following context and a question, generate an answer based on this context only.
In the answer try to provide as much text as possible from "response" section in the source document context without making much changes.
If the answer is not found in the context, kindly state "I don't know." Don't try to make up an answer.

CONTEXT: {context}

QUESTION: {question}"""

PROMPT = PromptTemplate(
    template=prompt_template, input_variables=["context", "question"]
)

chain_type_kwargs = {"prompt": PROMPT}

chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=retriever,
    input_key="query",
    return_source_documents=True,
    chain_type_kwargs=chain_type_kwargs
)
```

```python
response = chain.invoke({"query": "Do you provide job assistance and also do you provide job guarantee?"})
print(response["result"])  # The generated answer
print(response["source_documents"])  # The retrieved documents
```

```python
print(response)
```

**As you can see above, the answer of question comes from two different FAQs within our csv file and it is able to pull those questions and merge them nicely**
